```python
import numpy as np
import pandas as pd
np.random.seed(42)
num_samples = 100000


data = {
    #دما
    'temperature' : np.random.normal(loc=30 , scale= 5 , size = num_samples),
    #میزان بارش
    'precipitation' : np.random.normal(loc = 500 , scale= 100 , size = num_samples),
    #رطوبت
    'humidity': np.random.normal(loc=50 , scale= 10 , size = num_samples),
    'water_level':np.random.uniform(0,100,size=num_samples),
    'cultivated_area':np.random.uniform(0.1,0.9,size=num_samples)
}

df = pd.DataFrame(data)
df
```

```python
def risk(row):
    if row['temperature'] > 30 and row['precipitation'] < 200 and row['water_level'] < 0.2 :
        return 2 #پرخطر
    elif row['temperature'] > 30 or row['precipitation'] < 200 : 
        return 1 #کم خطر
    else : 
        return 0 #بدون خطر
    
    
```

```python
df['risk'] = df.apply(risk , axis = 1)
```

```python
df
```

```python
x=df.drop('risk',axis=1)
y=df['risk']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 150)
model.fit(x_train , y_train)
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
acc=accuracy_score(y_pred,y_test)
```
