```python
# library ---> کارهای خاص دستورات خاص
#pandas ----> csv , excel , جدول dataframe

import pandas as pd
df = pd.read_csv("spotify_songs.csv")
df
```

def ---> کار مشخص ---> a + b = c

```python
def spotify(top_n = 3): 
    artist = input("enter your favorite artist : ")
    genre = input("چه ژانری آهنگ گوش میدی :(pop , rap , edm , r&b , rock)")
    mood = input("چه مودی داری ( happy , sad , energetic , calm)")

    mood_dict ={
        'happy' : 'hapy' , 'energetic' : 'energetik' , 'happy' : 'شاد' 
        }
    
    mood = mood_dict.get(mood , mood)
    songs = df.copy()

    if artist : 
        #songs['Unnamed: 2'] ----> ستون ایم خواننده 
        #.str ----> تایپ ستون داده از جنس رشته یا string هست
        #.contain ----> شامل شدن
        # case : روی حروف کوچک یا بزرگ حساس adele or ADele
        # regex : amir - *amir - #amir - @amir ---> off
        songs = songs[songs['Unnamed: 2'].str.contains(artist , case = False , na = False , regex = False )]

    if genre : 
        songs = songs[songs['playlist_genre'].str.contains(genre , case = False , na = False , regex = False )]

    if mood == 'happy' :
        songs = songs[songs['valence'] > 0.7]

    elif mood == 'sad' : 
        songs = songs[songs['valence'] < 0.4]
    
    elif mood == 'calm':
        if 'energy' in songs.columns:
            songs = songs[(songs['valence'].between(0.4 , 0.7 , inclusive= 'both')) & (songs['energy'] < 0.4)]
        else : 
            songs = songs[songs['energy'] < 0.4]

    elif mood == 'energetic' : 
        if 'energy' in songs.columns : 
            songs = songs[songs['energy'] > 0.5]
        else : 
            songs = songs[songs['valence'] > 0.5]

    
    elif songs.empty : 
        print(" music not found ....")

    return songs.head(3)
```

```python
spotify(top_n = 3)
```
