import ----> فراخوانی کتابخانه

```python
import random

computer = random.randint(0 , 60)

adad = int(input("guess number :"))
if adad == computer :
  print(".....آفریننن درستهههههه")
elif adad < computer :
  print(" .....عدد بالاتر باید بگی")
elif adad > computer :
  print("عدد کوچکتر باید بگی....")
```

تا زمانی که کاربر  درست حدس بزنه ادامه بده

```python
import random

computer = random.randint(0 , 56)

while True :
  adad = int(input("guess number :"))
  if adad == computer :
    print(".....آفریننن درستهههههه")
    break
  elif adad < computer :
    print(" .....عدد بالاتر باید بگی")
  elif adad > computer :
    print("عدد کوچکتر باید بگی....")
```
