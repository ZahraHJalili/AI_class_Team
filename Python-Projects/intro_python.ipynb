if - elif - else ----> دستورات شرطی

if ---> شرط اول و همیشه باید باشه -- شرط داره

elif ----> شرط های بعدی و دوم به  --- شرط داره

else ---> هر چی نشد در نطر بگیر اخرین حالت --- حالت یا شرط نداره

```python
a = int(input("enter :"))
b = int(input("enter :"))

if a>b :
  print(a+b)
elif a<b :
  print(a*b)

elif a==b :
  print(a - b)

else :
  print(" ok ")
```

```python
score_1 = int(input("enter ur first score :"))
score_2 = int(input("enter ur second score :"))
score_3 = int(input("enter ur third score :"))


average_score = (score_1 + score_2 + score_3)/3
print("your average score is :",average_score)

if average_score >= 80 :
  print("good job")
elif average_score <= 50 :
  print("fail")
elif 50 < average_score < 80 :
  print("try more")
else : print("invalid score")
```
