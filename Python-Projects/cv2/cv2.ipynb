```python
import cv2 #پردازش تصویر و رنگ
import matplotlib.pyplot as plt #برای کشیدن پلات برای عکس ها 

image = cv2.imread("OIP.jpg") #خواندن عکس
plt.imshow(image) #نمایش عکس
```

```python
image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB) #تغییر رنگ
plt.imshow(image)
```

```python
def ishow(image):
    cv2.imshow('zahra' , image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
#انتخاب دوربین سیستم
cam = cv2.VideoCapture(0)
```

```python
_ ,img_webcam = cam.read()
```

```python
plt.imshow(img_webcam)
```

```python
img = img_webcam
```

```python
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
ishow(img)
```

```python
model_face = cv2.CascadeClassifier("haarcascade_frontalcatface.xml")
```

```python
face = model_face.detectMultiScale(img)
face
```

```python
x1 = face[0][0] #راس اول مربع
y1 = face[0][1] #راس دوم مربع
x2 = x1 + face[0][2] #راس سوم مربع
y2 = y1 + face[0][3] #راس چهارم مربع
```

RGB ----> 0 TO 255

```python
img = cv2.rectangle(img , (x1 , y1) , (x2 , y2) , (0 , 0 , 255) , 7)
plt.imshow(img)
```

```python
model_eye = cv2.CascadeClassifier("haarcascade_eye.xml")
```

```python
eye = model_eye.detectMultiScale(img)
eye
```

```python
#چشم اول
a1 = eye[0][0] #راس اول مربع
b1 = eye[0][1] #راس دوم مربع
a2 = a1 + eye[0][2] #راس سوم مربع
b2 = b1 + eye[0][3] #راس چهارم مربع

#چشم دوم
c1 = eye[1][0] #راس اول مربع
d1 = eye[1][1] #راس دوم مربع
c2 = c1 + eye[1][2] #راس سوم مربع
d2 = d1 + eye[1][3] #راس چهارم مربع
```

```python
#برای کادر چشو اول
img = cv2.rectangle(img , (a1 , b1) , (a2 , b2) , (210 , 0 , 76) , 6)

#برای کادر چشم دوم
img = cv2.rectangle(img , (c1 , d1) , (c2 , d2) , (0 , 0 , 76) , 6)
```

```python
plt.imshow(img)
```

تمرین  ----> با استفاده از مدل haaracascade smile , fullbody پروژه رو ترکیب و انجاب بده

```python
while True : 
    _ ,img_webcam = cam.read()
    img = img_webcam
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    face = model_face.detectMultiScale(img)
    x1 = face[0][0] #راس اول مربع
    y1 = face[0][1] #راس دوم مربع
    x2 = x1 + face[0][2] #راس سوم مربع
    y2 = y1 + face[0][3] #راس چهارم مربع
    img = cv2.rectangle(img , (x1 , y1) , (x2 , y2) , (0 , 0 , 255) , 7)
    cv2.imshow("z" , img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```
