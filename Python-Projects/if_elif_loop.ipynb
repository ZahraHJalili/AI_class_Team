```python
name = input("enter your name :")
s = 0
for i in range(3):
    n = float(input("enter your score:"))
    s = s + n

mean = s / 3
print(mean)
if mean >= 12 :
    print(name ,"is pass...")
else :
    print(name , "is failed!!")
```

```python
def calculator():
    a = float(input("عدد اول را وارد کنید :"))
    b = float(input("عدد دوم را وارد کتید :"))
    o = input("عملگر مورد نظر خورد را وارد کنید :")

    if o == '+':
        print(a + b)
    elif o == '-':
        print(a-b)
    elif o == '*':
        print(a * b)
    else :
        if b == 0 :
            print("error!!!!!")
        else : 
            print(a / b)

calculator()
```

```python
def calculator(a , b , o):
    if o == '+':
        print(a + b)
    elif o == '-':
        print(a-b)
    elif o == '*':
        print(a * b)
    else :
        if b == 0 :
            print("error!!!!!")
        else : 
            print(a / b)


a = float(input("عدد اول را وارد کنید :"))
b = float(input("عدد دوم را وارد کتید :"))
o = input("عملگر مورد نظر خورد را وارد کنید :")
calculator(a , b , o)
```
